# Read Data from JSON

```python
import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

# Adding New Product Into Inventory

```python
prod_code = input("Enter the unique code: ")
prod_name = input("Enter the name of the product: ")
prod_price = int(input("Enter the Price of the product: "))
prod_qty =  int(input("Enter the Quantity of the product: "))
prod_type = input("Enter the type of Product: ")
prod_Flav = input("Enter the flavour of the Product: ")


if(prod_code in record.keys()):
    record[prod_code]['Qty'] = record[prod_code]['Qty'] + prod_qty
    print("Existing Product Updated")
else:
    record[prod_code] = {'Name': prod_name,'Price':prod_price,'Qty':prod_qty,'Type':prod_type,'Flavour':prod_Flav}
    print("New Product Added in the Inventory")
    
    
js = json.dumps(record)

fd = open('record.json','w')
fd.write(js)
fd.close()
```

# Delete a product from Inventory

```python
prod_code = input("Enter the unique code: ")

if (prod_code in record.keys()):
    del record[prod_code]
    print("Product is now deleted from Inventory")
else:
    print("Product Doesn't Exists")
```
